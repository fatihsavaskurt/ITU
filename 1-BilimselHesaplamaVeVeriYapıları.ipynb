### Copy Kullanımı

--Aşağıdaki örnekte görüldüğü gibi, copy denerek aslında array bilgisayara kopyalanır. <p>
--a_sub ile kopyalanmış veride 0.satırdaki 0.sütununu 100 yapıyoruz. <p>
--a_sub 0,0 100 olurken a arrayindeki 0,0 değerini değiştirmemiş oldu. <p>

-- Numpy'da slice alındığı zaman kendisi otomatik kopyalamaz, avantajı büyük bir veri var ise kopyalamadan işlem yapılabilir. <p>

-- Pandasta da kopyalamama eğilimi var, inplace parametresi ile orijinalde değişiklik de yapılabiliyor.

```python
a = np.arange(20).reshape(4,5)
```

```python
a
```

```python
a[:2,2:]
```

```python
a_sub = a[:2,2:].copy()
```

```python
a_sub[0,0] = 100
```

```python
a_sub
```

```python
a
```
