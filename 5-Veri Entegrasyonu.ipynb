### Function Application and Mapping

* Apply tablonun kolonlarına çalışırken map tablonun satırlarına çalışır. Yani map hücre hücre çalışıyor, apply direkt tüm kolonlara.
* Her iki fonksiyon da dataframe veya series içerisinde uygulamalar yapmak için kullanılır.

```python
frame = pd.DataFrame(np.random.randint(0,10,(4,3)), columns= list('bde'),
                     index = ["Utah", "Ohio", "Texas", "Oregon"])
```

```python
frame
```

```python
frame.min()
```

```python
frame.max()
```

```python
f = lambda x: x.max() - x.min()
```

```python
frame.apply(f)
```

```python
## map
```

```python
myfunct = lambda x: x**2
```

```python
frame.map(myfunct)
```

```python
def f(x):
    if type(x) == str:
        return x.strip()
    else:
        return x
### Tüm hücrelere git, stringse strip koy değil ise hiçbir şey yapma.
```
