### Function Application and Mapping

```python
frame = pd.DataFrame(np.random.randint(0,10,(4, 3)), columns=list('bde'),
                         index=['Utah', 'Ohio', 'Texas', 'Oregon'])
```

```python
frame.min()
```

```python
frame.max()
```

#### Lambda

Tek satırda fonksiyon yazmak için kullanılır. X aldığı girdi daha sonrasında da max ve mini alıp çıkartacak. Apply metodunun içine fonksiyon koyulması gerekir. 

```python
f = lambda x: x.max() - x.min()
```

```python
frame.apply(f)
```

```python
def f(x):
    return pd.Series([x.min(), x.max()], index = ['min', 'max'])
```

```python
frame.apply(f)
```
